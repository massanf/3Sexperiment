```python
import matplotlib.pyplot as plt
import numpy as np
import math

with open("aaa.csv") as f:
    rawdata = f.readlines()

x = []
ch1 = []
ch2 = []
    
for i in range (0, len(rawdata)):
    x.append(float(rawdata[i].split(",")[0]))
    ch1.append(float(rawdata[i].split(",")[1]))
    ch2.append(float(rawdata[i].split(",")[2]))
    ch1[i] += 1.12
    ch2[i] += 1.12
    

plt.subplots(figsize=(8.0, 4.0), dpi=80)
#plt.plot(x,ch1,linewidth=1)
plt.scatter(x[92:98],ch2[92:98],linewidth=1)
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
plt.legend(["V1", "V2"], loc ="lower right")
plt.axhline(0,xmin=0,xmax=1,c='black',linewidth=0.3)
plt.savefig("figures/2b.png")
plt.xlim(-0.000001,0.00002)
plt.show()
```
