```python
import matplotlib.pyplot as plt
import numpy as np
import math

with open("aaa.csv") as f:
    rawdata = f.readlines()

x = []
ch1 = []
ch2 = []
    
for i in rawdata:
    x.append(float(i.split(",")[0]))
    ch1.append(float(i.split(",")[1]))
    ch2.append(float(i.split(",")[2]))

plt.subplots(figsize=(8.0, 4.0), dpi=80)
#plt.plot(x,ch1,linewidth=1)
plt.plot(x,ch2,linewidth=1)
plt.xlabel("Time [s]")
plt.ylabel("Voltage [V]")
plt.legend(["V1", "V2"], loc ="lower right")
plt.savefig("figures/2b.png")
```
