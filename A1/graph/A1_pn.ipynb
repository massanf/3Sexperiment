```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.figure(figsize=(9, 5), dpi=80)

with open("data_pn_a_normal.csv") as f:
    rawdata = f.readlines()

th_v = [i/100 for i in range (0, 90)]
v_t = 0.026
i_s = 0.001
th_i = [i_s * (np.exp(i/(2*v_t)) - 1) for i in th_v]

v = []
i = []
v_ir = []

for k in range (1, len(rawdata)):
    v.append(float(rawdata[k].split(",")[0]))
    i.append(float(rawdata[k].split(",")[1]))
    v_ir.append(float(rawdata[k].split(",")[3]))

i_corrected = []
for k in range (0, len(v)):
    i_corrected.append((i[k]*(10**-6)-(v[k]/(v_ir[k]*1000)))*(10**6))

plt.scatter(v,i_corrected, s=3, c="black")

#plt.plot(th_v,th_i)
plt.xlabel("Voltage [V]")
plt.ylabel("Current [μA]")
plt.ylim(-1000,10000)
#plt.yscale("log")
plt.savefig("figures/data_pn_a_normal_1.png")
```
